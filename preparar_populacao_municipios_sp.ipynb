# População por município (SP): média 2022–2025

Lê `csvs/populacao_municipio_SP.csv`, calcula a média da coluna `populacao` entre os anos 2022 e 2025 para cada município e grava `csvs/populacao_municipio_SP_media_2022_2025.csv` para uso em `analise_feminicidios.ipynb` (normalização por habitante).

```python
import pandas as pd

from municipio_sp import chave_municipio_oficial
```

```python
pop = pd.read_csv("csvs/populacao_municipio_SP.csv")
pop = pop.loc[pop["ano"].between(2022, 2025)]

pop_media = (
    pop.groupby(["id_municipio", "municipio"], as_index=False)["populacao"]
    .mean()
    .rename(columns={"populacao": "populacao_media_2022_2025"})
)
pop_media["municipio_chave"] = pop_media["municipio"].map(chave_municipio_oficial)

saida = "csvs/populacao_municipio_SP_media_2022_2025.csv"
pop_media.to_csv(saida, index=False, encoding="utf-8-sig")
print(f"Salvo: {saida} ({len(pop_media)} municípios)")
pop_media.head(10)
```
