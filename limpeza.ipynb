# Ententendo o que precisa filtrar

**Análise detalhada das colunas por ano**: número de linhas e colunas em cada ano, listando todas as colunas. Ajuda a identificar inconsistências estruturais entre os períodos.

```python
# Análise de Colunas: Comparação entre os anos
print("=" * 80)
print("COLUNAS POR ANO")
print("=" * 80)

years_data = {
    "2022": df_2022,
    "2023": df_2023,
    "2024": df_2024,
    "2025": df_2025
}

for year, df in years_data.items():
    print(f"\n{year}: {df.shape[0]} linhas, {df.shape[1]} colunas")
    print(f"Colunas: {sorted(df.columns.tolist())}")

#Pegando essencialmente todas as colunas, e vendo como elas diferem ao longo dos anos
```

Bem perceptível que os asno 2024 e 2025 tem as mesmas colunas, porem ainda assim com os nomes nao padronizados, enquanto os anos 2023 e 2022 tem essa semelhanca entre si

(Apesar de que na verdade tem algumas informacoes que uma ou outra nao tem)

Agora precisamos **Encontrar colunas que são iguais em todos os anos e colunas que diferem**. Mostra quais colunas estão faltando em cada período, ajudando a identificar o que precisa ser padronizado.

**Apresenta as duas opções de dataset final**: a versão com todas as colunas (df_total) e a versão limpa apenas com colunas completas (df_total_completo) recomendada para análises.

**Salva a documentação completa em um arquivo Markdown** (DOCUMENTACAO_COLUNAS.md) com todas as informações de mapeamento e análise de colunas para referência futura.

```python
# Encontrar colunas comuns e diferences
print("\n" + "=" * 80)
print("ANÁLISE DE DIFERENÇAS NAS COLUNAS")
print("=" * 80)

all_columns = {year: set(df.columns) for year, df in years_data.items()}

# Colunas comuns em todos os anos
common_columns = set.intersection(*all_columns.values())
print(f"\n Colunas presentes em TODOS os anos ({len(common_columns)} colunas):")
print(sorted(common_columns))

# Colunas por ano que podem estar faltando em outras
print("\n\n" + "-" * 80)
print("COLUNAS ESPECÍFICAS POR ANO (não presentes em todos):")
print("-" * 80)

for year in ["2022", "2023", "2024", "2025"]:
    unique_to_year = all_columns[year] - common_columns
    if unique_to_year:
        print(f"\n{year}: {unique_to_year}")
    else:
        print(f"\n{year}: Nenhuma coluna única")

# Colunas faltando por ano
print("\n\n" + "-" * 80)
print("COLUNAS FALTANDO EM CADA ANO:")
print("-" * 80)

for year in ["2022", "2023", "2024", "2025"]:
    missing = set()
    for other_year in ["2022", "2023", "2024", "2025"]:
        if year != other_year:
            missing.update(all_columns[other_year] - all_columns[year])
    
    if missing:
        print(f"\n{year} (faltam {len(missing)} colunas):")
        for col in sorted(missing):
            print(f"  - {col}")
```

**Identifica as relações entre colunas que têm a mesma função mas nomes diferentes** (por exemplo, "CIDADE" vs "NOME_MUNICIPIO"). Também encontra colunas completamente novas em 2024-2025.

```python
# Mapeamento de Colunas Equivalentes (mesma função, nomes diferentes)
print("\n" + "=" * 80)
print("MAPEAMENTO DE COLUNAS COM MESMA FUNÇÃO (nomes diferentes)")
print("=" * 80)

column_mapping = {
    "PERÍODO": ["DESCR_PERIODO", "DESC_PERIODO"],
    "TIPO LOCAL": ["DESCR_TIPOLOCAL", "DESCR_SUBTIPOLOCAL"],
    "MUNICÍPIO": ["CIDADE", "NOME_MUNICIPIO"],
    "CIRCUNSCRIÇÃO (acentuação)": ["NOME_DELEGACIA_CIRCUNSCRIÇÃO", "NOME_DELEGACIA_CIRCUNSCRICAO"],
}

for func, variants in column_mapping.items():
    print(f"\n{func}:")
    for col in variants:
        years_with_col = [year for year, cols in all_columns.items() if col in cols]
        print(f"  {col:40} → {', '.join(years_with_col)}")

# Colunas novass (2024+)
print("\n" + "=" * 80)
print("COLUNAS NOVAS EM 2024-2025 (não em 2022-2023):")
print("=" * 80)
new_cols_2024 = all_columns["2024"] - (all_columns["2022"] | all_columns["2023"])
print(f"\nNovas em 2024: {new_cols_2024}")

# Comparar 2024 vs 2025
print("\n" + "-" * 80)
print("Diferença entre 2024 e 2025:")
print("-" * 80)
only_2024 = all_columns["2024"] - all_columns["2025"]
only_2025 = all_columns["2025"] - all_columns["2024"]
print(f"Apenas em 2024: {only_2024}")
print(f"Apenas em 2025: {only_2025}")
```

**Aplica a padronização de nomes de colunas** usando o mapeamento identificado, renomeando colunas equivalentes para nomes consistentes em todos os anos. Depois reconstrói o dataframe total com os nomes padronizados.

```python
# SOLUÇÃO: Padronizar nomes de colunas e filtrar apenas colunas completas

# 1. Definir mapeamento completo de renomeação
rename_mapping = {
    # Período
    "DESCR_PERIODO": "DESC_PERIODO",
    
    # Tipo de Local (2025 trouxe de volta DESCR_TIPOLOCAL)
    # Vamos manter DESCR_SUBTIPOLOCAL como padrão
    
    # Município
    "CIDADE": "NOME_MUNICIPIO",
    
    # Acentuação (remover circunflexo para padronizar)
    "NOME_DELEGACIA_CIRCUNSCRIÇÃO": "NOME_DELEGACIA_CIRCUNSCRICAO",
    "NOME_DEPARTAMENTO_CIRCUNSCRIÇÃO": "NOME_DEPARTAMENTO_CIRCUNSCRICAO",
    "NOME_MUNICIPIO_CIRCUNSCRIÇÃO": "NOME_MUNICIPIO_CIRCUNSCRICAO",
    "NOME_SECCIONAL_CIRCUNSCRIÇÃO": "NOME_SECCIONAL_CIRCUNSCRICAO",
}

# 2. Aplicar renomeação em cada dataframe
for year in ["2022", "2023", "2024", "2025"]:
    years_data[year] = years_data[year].rename(columns=rename_mapping)

# 3. Recriar df_total com colunas renomeadas
df_total = pd.concat([years_data[year] for year in ["2022", "2023", "2024", "2025"]], ignore_index=True)

print("=" * 80)
print("COLUNAS APÓS PADRONIZAÇÃO")
print("=" * 80)

for year in ["2022", "2023", "2024", "2025"]:
    print(f"\n{year}: {len(years_data[year].columns)} colunas")
    print(f"Colunas: {sorted(years_data[year].columns.tolist())}")
```

**Identifica colunas "completas" (presentes em todos os 4 anos) vs "incompletas"** (que faltam em algum período). Cria um dataframe filtrado com apenas as colunas completas para análises sem valores faltantes (NaN).

```python
# Identificar colunas COMPLETAS (presentes em todos os 4 anos)
print("\n" + "=" * 80)
print("ANÁLISE FINAL: COLUNAS COMPLETAS vs INCOMPLETAS")
print("=" * 80)

# Atualizar sets de colunas
all_columns_after = {year: set(years_data[year].columns) for year in ["2022", "2023", "2024", "2025"]}

# Colunas comuns em todos
complete_columns = set.intersection(*all_columns_after.values())
print(f"\n✓ COLUNAS COMPLETAS (presentes em todos 4 anos): {len(complete_columns)}")
print(sorted(complete_columns))

# Colunas incompletas
incomplete_columns = set.union(*all_columns_after.values()) - complete_columns
print(f"\n✗ COLUNAS INCOMPLETAS (faltam em algum ano): {len(incomplete_columns)}")

for col in sorted(incomplete_columns):
    years_present = [year for year in ["2022", "2023", "2024", "2025"] if col in all_columns_after[year]]
    years_missing = [year for year in ["2022", "2023", "2024", "2025"] if col not in all_columns_after[year]]
    print(f"  {col:45} → Tem: {years_present}, Falta em: {years_missing}")

# Criar df_total_completo com apenas colunas completas
df_total_completo = df_total[sorted(complete_columns)]

print("\n" + "=" * 80)
print("RESULTADO FINAL")
print("=" * 80)
print(f"\ndf_total (com todas as colunas):")
print(f"  - Linhas: {df_total.shape[0]}")
print(f"  - Colunas: {df_total.shape[1]}")

print(f"\ndf_total_completo (apenas colunas presentes em todos os anos):")
print(f"  - Linhas: {df_total_completo.shape[0]}")
print(f"  - Colunas: {df_total_completo.shape[1]}")
print(f"  - Colunas: {sorted(df_total_completo.columns.tolist())}")
```

```python
# Comparação antes/depois e duas opções de dataset final
print("\n" + "=" * 80)
print("DUAS OPCOES PARA O DATASET FINAL")
print("=" * 80)

print(f"\n📊 OPTION 1: DATA FRAME TOTAL (COM NOMES PADROES DE COLUNAS)")
print(f"   Colunas: {df_total.shape[1]}")
print(f"   Linhas: {df_total.shape[0]:,}")
print(f"   ⚠️  Contém valores NaN onde colunas faltam:")
for col in sorted(incomplete_columns):
    null_count = df_total[col].isna().sum() if col in df_total.columns else 0
    print(f"      - {col:40} → {null_count:,} NaN values")

print(f"\n📊 OPCAO 2: df_total_completo (only complete columns)")
print(f"   Colunas: {df_total_completo.shape[1]}")
print(f"   Linhas: {df_total_completo.shape[0]:,}")
print(f"   ✓ SEM valores NaN (dados 100% completos)")
print(f"   ✓ {len(complete_columns)} colunas consistentes em todos os anos")

print("\n" + "-" * 80)
print("RECOMENDAÇÃO:")
print("-" * 80)
print("Use 'df_total_completo' para análises mais limpas e consistentes!")
print("Todas as colunas têm dados em todos os 4 anos de períodos.")

# Preview dos dados
print("\n" + "=" * 80)
print("PREVIEW: df_total_completo.head()")
print("=" * 80)
print(df_total_completo.head())
```
