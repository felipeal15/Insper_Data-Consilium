```python
import requests
import json
from time import sleep
import pandas as pd
```

```python
request = requests.get('https://www.ssp.sp.gov.br/v1/ViolenciaMulher/RecuperaDadosPorAno?ano=2023').json()
```

```python
data_anos = {}
for i in range(2018,2027):
    data_anos[str(i)] = requests.get(f'https://www.ssp.sp.gov.br/v1/ViolenciaMulher/RecuperaDadosPorAno?ano={i}').json()['data']
    sleep(.1)
```

Para selecionar mais crimes, adicionar no dicionário `CODIGOS`.

```python
CODIGOS = {
    134: 'FEMINICÍDIO',
    # 135: 'HOMICÍDIO DOLOSO - TOTAL',
    # 58: 'HOMICÍDIO DOLOSO (exclui FEMINICÍDIO)',
    # 59: 'HOMICÍDIO CULPOSO',
    # 60: 'TENTATIVA DE HOMICÍDIO',
    # 61: 'LESÃO CORPORAL DOLOSA',
    # 62: 'MAUS TRATOS',
    # 63: 'CALÚNIA - DIFAMAÇÃO - INJÚRIA',
    # 64: 'CONSTRANGIMENTO ILEGAL',
    # 65: 'AMEAÇA',
    # 66: 'INVASÃO DE DOMICÍLIO',
    # 67: 'DANO',
    # 68: 'ESTUPRO CONSUMADO',
    # 69: 'ESTUPRO TENTADO',
    # 70: 'ESTUPRO DE VULNERÁVEL CONSUMADO',
    # 136: 'ESTUPRO DE VULNERÁVEL TENTADO',
    # 71: 'OUTROS C/C/ DIGNIDADE SEXUAL'
}
```

```python

db = []
for i in range(2018,2027):
    for mensal in data_anos[str(i)]:
        for delito in mensal['dadosMes']:
            
            if delito['idDelito'] in CODIGOS.keys(): # id feminicídio
                linha = {
                    'ano': i,
                    'mes': mensal['mes'],
                    'tipo': CODIGOS[delito['idDelito']],
                    'capital': delito['capital'] ,
                    'demacro': delito['demacro'] ,
                    'interior': delito['interior'] ,
                    'total': delito['total'],
                }
                db.append(linha)
                break
```

## Banco de dados:

```python
db = pd.DataFrame(db)
display(db)
```
